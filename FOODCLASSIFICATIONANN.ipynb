FOOD CLASSIFICATION USING ARTIFICIAL NEURAL NETWORK IN DEEP LEARNING


```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("food classification.csv")
```

```python
df
```

```python
df=df.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
df
```

```python
x=df.drop("class",axis=1)
y=df["class"]
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
y1=lb.fit_transform(y)
y=pd.get_dummies(y).values
```

```python
y1
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=46)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

```python
import tensorflow as tf
import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from scikeras.wrappers import KerasClassifier,KerasRegressor
```

```python
classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(Dense(units= 30,kernel_initializer='glorot_uniform',activation='ReLU',input_dim = 17))


# Adding the second hidden layer
classifier.add(Dense(units = 13,kernel_initializer='glorot_uniform',activation='ReLU'))
classifier.add(Dense(units = 14,kernel_initializer='glorot_uniform',activation='ReLU'))


# Adding the output layer
classifier.add(Dense(units = 10,kernel_initializer='glorot_uniform',activation='ReLU'))



# Adding the output layer
classifier.add(Dense(units= 3, activation = 'softmax'))

# Compiling the ANN
classifier.compile(optimizer = 'Adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])

# Fitting the ANN to the Training set
model_history=classifier.fit(x_train, y_train,validation_split=0.33, batch_size = 10, epochs = 100)
```

```python
y_pred=classifier.predict(x_test)
y_test_class=np.argmax(y_test,axis=1)
y_pred_class=np.argmax(y_pred,axis=1)
y_pred_class
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test_class,y_pred_class))
print(confusion_matrix(y_test_class,y_pred_class))
```

```python
def create_model(neurons=30):
  model =Sequential()
  model.add(Dense(units=neurons,activation='relu',input_dim = 17))
  model.add(Dense(units = 13,activation='relu'))
  model.add(Dense(units = 14,activation='relu'))
  model.add(Dense(units = 10,activation='relu'))
  model.add(Dense(units = 3,activation='Softmax'))
  model.compile(optimizer = 'Adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
  return model
```

```python
base_model=KerasClassifier(build_fn=create_model,verbose=0,batch_size=10,epochs=100,neurons=30)
base_model.fit(x_train,y_train,epochs=100)
```

```python
best_score=base_model.score(x_test,y_test)
best_score
```

```python
y_pred = base_model.predict(x_test)

# Ensure y_pred is in the correct format for classification_report
y_pred1 = np.argmax(y_pred, axis=1)  # Convert one-hot encoded predictions to class labels

# Get the classification report
```

```python
print(classification_report(y_test_class,y_pred1))
print(confusion_matrix(y_test_class,y_pred1))
```

Hyperparameter Tuning using RandomizedSearchCV FOR ANN


```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
batch_size=[10,20,30,40,60,80,100]
epochs=[1,3,10,20,50,100,150]
neurons =[1,10,20,30,40]
params=dict(batch_size=batch_size,epochs=epochs,neurons=neurons)
params
```

```python
search=RandomizedSearchCV(estimator=base_model,param_distributions=params,cv=3)
```

```python
search_result=search.fit(x_train,y_train)
```
